Začínáme importováním potřebných balíčků a načtením datasetu:

````python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
````

````python
data = pd.read_csv('data.csv')
````

Podívejme se, jaké příznaky máme, jakého jsou typu a kolik unikátních hodnot má každý příznak:

````python
data.head()
````

````python
display(data.dtypes)
data.nunique()  # ignoruje NaNs
````

Některé sloupce nepotřebujeme, protože opravdu nemají vliv na přežití pasažérů. Můžeme je odstranit:

````python
data.drop(columns=['ID','name', 'ticket', 'home.dest'], inplace=True)
````

Teď můžeme zobrazit, co nám zbývá:

````python
data.head()
````

Na přežití pasažérů mají vliv taková data, jako je věk pasažéra, pohlaví, bohatství (což nám sděluje pclass, možná fare, tj. cena lístku), přítomnost příbuzných na palubě (sibsp, parch). Ponecháme také místo nalodění (embarked).

Příznak čísla kajuty (cabin) se mi nezdá příliš užitečný, ale v těchto číslech se vyskytují i písmenka. Předpokládám, že tato písmenka mohou značit rozmístění kajuty vzhledem k části lodi (příď, záď). Tato písmenka přidáme do nového příznaku cabin_class, a příznak cabin zahodíme:

````python
data['cabin_class'] = data['cabin'].apply(lambda x: x[0] if pd.notnull(x) else np.nan)
````

````python
data.drop(columns=['cabin'], inplace=True)
````

````python
data.dtypes
````

Nyní převedeme některé příznaky na kategorický typ:

````python
data['sex'] = data['sex'].astype("category")
data['embarked'] = data['embarked'].astype("category")
data['cabin_class'] = data['cabin_class'].astype("category")
````

````python
data.head()
````

````python
data.dtypes
````

Nyní můžeme nahradit všechny kategorické příznaky číselnými kódy:

````python
data.sex = data.sex.cat.codes
data.embarked = data.embarked.cat.codes
data.cabin_class = data.cabin_class.cat.codes
````

````python
data.dtypes
````

S chybějícími hodnotami se vypořádáme nejjednodušším způsobem - nahradíme je unikátní hodnotou -1:

````python
data = data.replace(np.nan, -1)
````

````python
data.head()
````

Nyní můžeme přistoupit k rozdělení dat na trénovací (60%), validační (20%) a testovací (20%) části:

````python
Xdata = data.drop("survived", axis = 1)
ydata = data.survived
````

````python
rd_seed = 333 

Xtrain, Xtest, ytrain, ytest = train_test_split(Xdata, ydata, test_size=0.4, random_state=rd_seed) 
display(Xtrain.shape)
display(ytrain.shape)
display(Xtest.shape)
display(ytest.shape)
````

````python
Xval, Xtest, yval, ytest = train_test_split(Xtest, ytest, test_size=0.5, random_state=rd_seed) 
display(Xval.shape)
display(yval.shape)
display(Xtest.shape)
display(ytest.shape)
````

### Implementace stromu

Pro implementaci stromu použijeme balíček `tree` knihovny `scikit-learn`.

````python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 1)
clf.get_params()
````

Použijeme metodu fit(), kde model naučíme.

````python
clf.fit(Xtrain, ytrain)
````

````python
clf = DecisionTreeClassifier(max_depth=50)
clf.fit(Xtrain, ytrain)
````

Nyní můžeme použít metodu predict(), abychom získali predikce, které dává strom pro trénovací data.

````python
ypredicted = clf.predict(Xtrain)
ypredicted
````

````python
(ytrain - ypredicted).value_counts(normalize=True)
````

Zkontrolujeme přesnost dat:

````python
import sklearn.metrics as metrics
print('accuracy score (train): {0:.6f}'.format(metrics.accuracy_score(ytrain, clf.predict(Xtrain))))
````

````python
print('accuracy score (test): {0:.6f}'.format(metrics.accuracy_score(ytest, clf.predict(Xtest))))
````

#### Ladění hyperparameterů 

Začněme s hyperparametrem max_depth. Prozkoumáme hodnoty od 1 do 121.

````python
val_acc = []
train_acc = []
depths = range(1,121)
for depth in depths:
    clf = DecisionTreeClassifier(max_depth=depth)
    clf.fit(Xtrain, ytrain)
    train_acc.append(metrics.accuracy_score(ytrain, clf.predict(Xtrain)))
    val_acc.append(metrics.accuracy_score(yval, clf.predict(Xval)))
````

````python

%matplotlib inline
plt.figure(figsize=(20,6))
plt.plot(depths, train_acc,'or-')
plt.plot(depths, val_acc,'ob-')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend(['train', 'validation']);
````

````python
optimal_depth = np.argmax(val_acc) + 1
optimal_depth
````

Spočítejme testovací accuracy pro nejlepší nalezenou hodnotu max_depth:

````python
clf = DecisionTreeClassifier(max_depth=optimal_depth)
clf.fit(Xtrain, ytrain)
print('accuracy score (test): {0:.6f}'.format(metrics.accuracy_score(ytest, clf.predict(Xtest))))
````

#### Ladění více hyperparameterů

 použijeme ParameterGrid balíček.

````python
from sklearn.model_selection import ParameterGrid
param_grid = {
    'max_depth': range(1,50), 
    'criterion': ['entropy', 'gini']
}
param_comb = ParameterGrid(param_grid)
# just to see the result:
[param_comb[i] for i in range(2)]
````

````python
val_acc = []
train_acc = []
for params in param_comb:
    clf = DecisionTreeClassifier(max_depth=params['max_depth'], criterion=params['criterion'])
    clf.fit(Xtrain, ytrain)
    train_acc.append(metrics.accuracy_score(ytrain, clf.predict(Xtrain)))
    val_acc.append(metrics.accuracy_score(yval, clf.predict(Xval)))
````

````python
import matplotlib
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
plt.plot(train_acc,'or-')
plt.plot(val_acc,'ob-')
plt.xlabel('hyperparametr index')
plt.ylabel('accuracy')
plt.legend(['train', 'validation']);
````

````python
best_params = param_comb[np.argmax(val_acc)]
best_params
````

````python
clf = DecisionTreeClassifier(**best_params)
clf.fit(Xtrain, ytrain)
print('accuracy score (train): {0:.6f}'.format(metrics.accuracy_score(ytrain, clf.predict(Xtrain))))
print('accuracy score (validation): {0:.6f}'.format(metrics.accuracy_score(yval, clf.predict(Xval))))
print('accuracy score (test): {0:.6f}'.format(metrics.accuracy_score(ytest, clf.predict(Xtest))))
````

Dostali jsme výsledek pro rozhodovací strom, můžeme přejít k metodě nejbližších sousedů.

### KNN
Nyní použijeme `KNeighborsClassifier` z knihovny `sklearn`.

````python
from sklearn.neighbors import KNeighborsClassifier

clfKNN = KNeighborsClassifier(n_neighbors =5)
clfKNN.fit(Xtrain, ytrain)
print(f"Validation accuracy: {metrics.accuracy_score(yval, clfKNN.predict(Xval)):0.4f}")
````

 Výsledek není až tak moc dobrý. Pojďme si vyzkoušet standardizaci jako typ normalizace.

````python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)

Xval_scaled = scaler.transform(Xval)
Xtest_scaled = scaler.transform(Xtest)

clfKNN = KNeighborsClassifier()
clfKNN.fit(Xtrain_scaled, ytrain)
print(f"Validation accuracy of normalized model: {clfKNN.score(Xval_scaled, yval):0.4f}")
````

Zkusíme si tedy nyní **min-max normalizaci**. To uděláme pomocí třídy `MinMaxScaler` z modulu `sklearn.preprocessing`.

````python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)

Xval_scaled = scaler.transform(Xval)
Xtest_scaled = scaler.transform(Xtest)
clfKNN = KNeighborsClassifier()
clfKNN.fit(Xtrain_scaled, ytrain)
print(f"Validation accuracy of normalized model: {clfKNN.score(Xval_scaled, yval):0.4f}")
````

Min-max normalizace dala lepší výsledek, proto použijeme právě tento způsob normalizace.

Podívejme se nyní na vliv hyperparametrů na výkonnost predikce.

````python
val_acc = []
train_acc = []

kneighbors = {
    'n_neighbors':range(1,10),
    'p': range(1,10)
}
kneighbors = range(3,40)
for k in kneighbors:
    clfKNN = KNeighborsClassifier(n_neighbors = k)
    clfKNN.fit(Xtrain_scaled, ytrain)
    val_acc.append(clfKNN.score(Xval_scaled, yval))
    train_acc.append(clfKNN.score(Xtrain_scaled, ytrain))
````

````python
%matplotlib inline
plt.figure(figsize=(20,6))
plt.plot(kneighbors, train_acc,'or-')
plt.plot(kneighbors, val_acc,'ob-')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.legend(['train', 'validation']);
````

````python
clfKNN.fit(Xtrain_scaled, ytrain)
print(f"Validation accuracy of normalized model: {clfKNN.score(Xval_scaled, yval):0.4f}")
````

Zkusila jsem nastavit parametr k a dostala jsem hodnotu trochu lepší.

### Porovnání výsledků

````python
print("Rozhodovací strom")
print('Accuracy score (train): {0:.6f}'.format(metrics.accuracy_score(ytrain, clf.predict(Xtrain))))
print('Accuracy score (test): {0:.6f}'.format(metrics.accuracy_score(ytest, clf.predict(Xtest))))

print("\nk nejbližších sousedů")
print('Accuracy score (train): {0:.6f}'.format(metrics.accuracy_score(ytrain, clfKNN.predict(Xtrain_scaled))))
print('Accuracy score (test): {0:.6f}'.format(metrics.accuracy_score(ytest, clfKNN.predict(Xtest_scaled))))
````

### Matice záměn

Spočítáme testovací matici zmáněn pro oba modely.

````python
cfDT = metrics.confusion_matrix(ytest, clf.predict(Xtest))
print("Rozhodovací strom")
print(cfDT)
cfKNN = metrics.confusion_matrix(ytest, clfKNN.predict(Xtest_scaled))
print("\nk nejbližších sousedů")
print(cfKNN)
````

Teď spočítáme TPR, FPR, FNR a TNR. Dále určíme recall, precision a hlavně spočítáme accuracy a F1.
Porovnejme oba modely.

````python
def print_metrics(cf, ytest, ypred):
    cf_r = np.dot(np.diag(1./cf.sum(axis = 1)), cf)
    cf_r = metrics.confusion_matrix(ytest, ypred, normalize = 'true')
    print(cf_r)

    print(f"\nTPR = {cf_r[1,1]:.4f}\nFPR = {cf_r[0,1]:.4f}\nFNR = {cf_r[1,0]:.4f}\nTNR = {cf_r[0,0]:.4f}")
    cf_rp = metrics.confusion_matrix(ytest, ypred, normalize = 'pred')
    print(f"\nrecall = TPR = {cf_r[1,1]:.4f}, precision = {cf_rp[1,1]:.4f}")
    
    print(f'Accuracy score (test): {metrics.accuracy_score(ytest, ypred):.4f}')
    print(f'F1 score (test): {metrics.f1_score(ytest, ypred):.4f}')
    
print("==========================")
print("Rozhodovací strom")
print_metrics(cfDT, ytest, clf.predict(Xtest))

print("==========================")
print("k nejbližších sousedů")
print_metrics(cfKNN, ytest, clfKNN.predict(Xtest_scaled))
````

Hodnoty F1 a Accuracy score jsou celkem blízko u sebe v obou případech, u KNN je trochu lepší. Zkusíme se podívat i na ROC křivku.

### ROC křivka a AUC

Teď nakresleme ROC křivku pro oba případy.

````python
y_pred_proba = clf.predict_proba(Xtest)
p1_hat = y_pred_proba[:,1]
````

````python
fpr, tpr, thresholds = metrics.roc_curve(ytest, p1_hat)

AUC = metrics.roc_auc_score(ytest, y_pred_proba[:,1])

plt.figure()
plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
plt.plot(fpr, tpr, color="darkorange", lw=2, label = f"ROC křivka (AUC = {AUC:.3f})")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC křivka")
plt.legend(loc="lower right");
````

````python
y_pred_proba = clfKNN.predict_proba(Xtest_scaled)
fpr, tpr, thresholds = metrics.roc_curve(ytest, y_pred_proba[:,1])

AUC = metrics.auc(fpr, tpr)

plt.figure()
plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
plt.plot(fpr, tpr, color="darkorange", lw=2, label = f"ROC křivka (AUC = {AUC:.3f})")
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC křivka")
plt.legend(loc="lower right")
plt.show()
````

Ze všech porovnání nám vychází, že model KNN je v tomto případě lepší.
Takže použijeme tento model pro další testování.
